```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
detector_file = "../../configs/detector.csv"
df = pd.read_csv(detector_file)
```

```python
cx, cy, cz = df["cx"], df["cy"], df["cz"]
```

```python
cr = np.sqrt(cx**2 + cy**2)
cR = np.sqrt(cx**2 + cy**2 + cz**2)
```

Remove `cr=0` and `cR=0` from the inputs.

```python
cr.shape, cr[cr < 0.1].shape, cr[cr < 0.1].shape[0] / cr.shape[0]
```

```python
cr_clean = cr[cr > 0.1]
cz_clean = cz[cr > 0.1]
cx_clean = cx[cr > 0.1]
cy_clean = cy[cr > 0.1]
```

```python
plt.scatter(cz_clean, cr_clean, s=0.5)
plt.xlabel("z (mm)")
plt.ylabel("r (mm)")
plt.tight_layout()
plt.savefig("detector_module_center_rz.png")
```

```python
plt.scatter(cx_clean, cy_clean, s=0.5)
plt.xlabel("x (mm)")
plt.ylabel("y (mm)")
plt.tight_layout()
plt.savefig("detector_module_center_xy.png")
```
